-> Soluções mais aptas têm uma maior probabilidade de serem selecionadas, mas soluções menos aptas também têm uma chance de serem escolhidas para manter diversidade na população.

CROSSOVER, MUTAÇÃO E GERAÇÃO DE 2000

```python
# Função para ler o arquivo TSP e extrair as coordenadas das cidades
def ler_arquivo_tsp(nome_arquivo):
    pontos = []
    with open(nome_arquivo, 'r') as arquivo:
        lendo_pontos = False
        for linha in arquivo:
            linha = linha.strip()
            if lendo_pontos:
                if linha == "EOF":
                    break
                numero, x, y = map(float, linha.split())
                pontos.append((x, y))
            elif linha == "NODE_COORD_SECTION":
                lendo_pontos = True
    return pontos

import random

# Função para calcular a distância entre duas cidades
def calcular_distancia(cidade1, cidade2):
    x1, y1 = cidade1
    x2, y2 = cidade2
    distancia = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return distancia

# Função para calcular a distância total de um caminho
def calcular_distancia_total(caminho, pontos):
    distancia_total = 0
    for i in range(len(caminho) - 1):
        cidade_atual = pontos[caminho[i]]
        proxima_cidade = pontos[caminho[i + 1]]
        distancia_total += calcular_distancia(cidade_atual, proxima_cidade)
    # Adicione a distância da última cidade de volta à primeira cidade para formar um ciclo
    distancia_total += calcular_distancia(pontos[caminho[-1]], pontos[caminho[0]])
    return distancia_total

# Função para inicializar a população com 50 indivíduos
def inicializar_populacao(pontos, tamanho_populacao):
    populacao = []
    for _ in range(tamanho_populacao):
        individuo = random.sample(list(range(len(pontos))), len(pontos))  # Embaralha as cidades
        populacao.append(individuo)
    return populacao

# Função de aptidão
def calcular_aptidao(individuo, pontos):
    distancia = calcular_distancia_total(individuo, pontos)
    return 1 / distancia  # Quanto menor a distância, maior a aptidão

# Função para seleção por torneio
def selecao_por_torneio(populacao, aptidoes, tamanho_torneio):
    pais_selecionados = []
    for _ in range(len(populacao)):
        torneio = random.sample(range(len(populacao)), tamanho_torneio)
        vencedor_torneio = max(torneio, key=lambda x: aptidoes[x])
        pais_selecionados.append(populacao[vencedor_torneio])
    return pais_selecionados

# Função para realizar o crossover (recombinação) entre dois pais
def crossover(pai1, pai2):
    ponto_corte = random.randint(1, len(pai1) - 1)
    filho1 = pai1[:ponto_corte] + [gene for gene in pai2 if gene not in pai1[:ponto_corte]]
    filho2 = pai2[:ponto_corte] + [gene for gene in pai1 if gene not in pai2[:ponto_corte]]
    return filho1, filho2

# Função para realizar a mutação de um indivíduo
def mutacao(individuo, taxa_mutacao):
    if random.random() < taxa_mutacao:
        indice1, indice2 = random.sample(range(len(individuo)), 2)
        individuo[indice1], individuo[indice2] = individuo[indice2], individuo[indice1]
    return individuo

# Nome do arquivo TSP
nome_arquivo = 'xqf131.tsp'

# Ler as coordenadas dos pontos
pontos = ler_arquivo_tsp(nome_arquivo)

# Inicializar a população com 50 indivíduos
tamanho_populacao = 50
populacao_inicial = inicializar_populacao(pontos, tamanho_populacao)

# Número de gerações desejadas
numero_geracoes = 2000

# Tamanho do torneio (quantos indivíduos competirão)
tamanho_torneio = 5

# Taxa de mutação (probabilidade de ocorrer mutação em um indivíduo)
taxa_mutacao = 0.1

# Loop para realizar as gerações
for geracao in range(numero_geracoes):
    # Calcular a aptidão para cada indivíduo na população
    aptidoes = [calcular_aptidao(individuo, pontos) for individuo in populacao_inicial]

    # Realizar a seleção por torneio
    pais_selecionados = selecao_por_torneio(populacao_inicial, aptidoes, tamanho_torneio)

    # Realizar o crossover entre os pais selecionados e aplicar mutação
    descendentes = []
    for i in range(0, len(pais_selecionados), 2):
        pai1 = pais_selecionados[i]
        pai2 = pais_selecionados[i + 1]
        filho1, filho2 = crossover(pai1, pai2)
        filho1 = mutacao(filho1, taxa_mutacao)
        filho2 = mutacao(filho2, taxa_mutacao)
        descendentes.append(filho1)
        descendentes.append(filho2)

    # Substituir a população antiga pela nova população de descendentes
    populacao_inicial = descendentes

    # Calcular e imprimir a melhor aptidão da geração atual
    melhor_aptidao = max(aptidoes)
    print(f"Geração {geracao + 1} - Melhor Aptidão: {melhor_aptidao:.6f}")

# Encontrar o melhor indivíduo da última geração
melhor_individuo = populacao_inicial[aptidoes.index(max(aptidoes))]

# Imprimir a ordem das cidades no melhor indivíduo
print("Melhor Caminho Encontrado:")
print(melhor_individuo)
```
